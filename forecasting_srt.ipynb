```python
import pandas as pd
from prophet import Prophet
import matplotlib.pyplot as plt

# Leer el archivo CSV
df = pd.read_csv('cuadro1.csv', delimiter=';')

# Limpiar y transformar los datos
df = df.melt(id_vars=['Jurisdiccin'], var_name='ao', value_name='valor')
df['ao'] = pd.to_datetime(df['ao'], format='%Y')

# Funcin para predecir con Prophet y generar grficos
def predict_with_prophet(df, jurisdiccion):
    # Filtrar los datos por jurisdiccin
    df_jurisdiccion = df[df['Jurisdiccin'] == jurisdiccion].dropna()
    
    # Preparar los datos para Prophet
    df_jurisdiccion = df_jurisdiccion.rename(columns={'ao': 'ds', 'valor': 'y'})

    # Crear y ajustar el modelo
    model = Prophet()
    model.fit(df_jurisdiccion)

    # Crear un DataFrame con las fechas futuras
    future = model.make_future_dataframe(periods=4, freq='Y')

    # Realizar la prediccin
    forecast = model.predict(future)

    # Graficar los resultados
    fig, ax = plt.subplots(figsize=(10, 6))
    model.plot(forecast, ax=ax)
    ax.set_title(f'Prediccin para la jurisdiccin {jurisdiccion}')
    ax.set_xlabel('Ao')
    ax.set_ylabel('Valor')
    plt.show()

    return forecast

# Predecir y graficar para cada jurisdiccin
for jurisdiccion in df['Jurisdiccin'].unique():
    forecast = predict_with_prophet(df, jurisdiccion)
    print(f'Predicciones para la jurisdiccin {jurisdiccion}:\n', forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(3))
```

```python
import pandas as pd
from prophet import Prophet
import matplotlib.pyplot as plt
import numpy as np
np.float_ = np.float64

# Leer el archivo CSV con el delimitador ;
df = pd.read_csv('cuadro1.csv', delimiter=';')


# Limpiar y transformar los datos
df = df.melt(id_vars=['Jurisdiccin'], var_name='ao', value_name='valor')
df['ao'] = pd.to_datetime(df['ao'], format='%Y')

# Copiar los valores de 2019 a 2020
for provincia in df['Jurisdiccin'].unique():
    df_2019 = df[(df['Jurisdiccin'] == provincia) & (df['ao'].dt.year == 2019)]
    df_2020 = df[(df['Jurisdiccin'] == provincia) & (df['ao'].dt.year == 2020)]
    df.loc[df_2020.index, 'valor'] = df_2019['valor'].values

# Funcin para predecir con Prophet y generar grficos
def predict_with_prophet(df, jurisdiccion):
    # Filtrar los datos por jurisdiccin
    df_jurisdiccion = df[df['Jurisdiccin'] == jurisdiccion].dropna()
    
    # Preparar los datos para Prophet
    df_jurisdiccion = df_jurisdiccion.rename(columns={'ao': 'ds', 'valor': 'y'})

    # Crear y ajustar el modelo
    model = Prophet()
    model.fit(df_jurisdiccion)

    # Crear un DataFrame con las fechas futuras
    future = model.make_future_dataframe(periods=4, freq='Y')

    # Realizar la prediccin
    forecast = model.predict(future)

    # Graficar los resultados
    fig, ax = plt.subplots(figsize=(10, 6))
    model.plot(forecast, ax=ax)
    ax.set_title(f'Prediccin para la jurisdiccin {jurisdiccion}')
    ax.set_xlabel('Ao')
    ax.set_ylabel('Valor')
    plt.show()

    return forecast

# Predecir y graficar para cada jurisdiccin
for jurisdiccion in df['Jurisdiccin'].unique():
    forecast = predict_with_prophet(df, jurisdiccion)
    print(f'Predicciones para la jurisdiccin {jurisdiccion}:\n', forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(3))
```
